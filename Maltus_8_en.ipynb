## 1. Thomas Robert Malthus: Personality, Character, Intellectual Portrait

Thomas Robert Malthus (1766–1834) remains a polarizing figure: from "the man we love to hate" to a prophet of demographic crises [[Cambridge]]. He was the seventh child in a wealthy family; his father, Daniel Malthus, was an enlightened gentleman personally acquainted with Jean-Jacques Rousseau and David Hume, shaping Thomas's intellectual upbringing [[5]].

**Personal Traits:**
*   **Appearance & Manners:** Contemporaries described Malthus as "a handsome man with an aristocratic nose, sharp eyes, and a high forehead," dressed as a gentleman of his era, with short curly hair [[1]].
*   **Character:** Retrospective psychological assessments suggest an **INTJ** personality type (Introverted, Intuitive, Thinking, Judging) — a strategic thinker inclined toward systemic analysis and long-term forecasting [[2]].
*   **Temperament:** Despite the pessimism of his conclusions, Malthus was personally known as gentle, polite, and respected in conversation. He avoided long travels, preferring to "journey through books" — his library at Jesus College Cambridge contained Captain Cook's journals, Jesuit missionary chronicles, Benjamin Franklin's essays, enabling him to construct a *global* demographic theory, not merely a European one [[Cambridge]].
*   **Professional Position:** For most of his life, Malthus was not just a clergyman but Professor of Political Economy at the East India Company College in Haileybury, training clerks for British imperial administration. This placed him at the heart of imperial discourse and gave him access to global demographic data [[Cambridge]].

**Intellectual Courage:**
Malthus was among the few thinkers of his time to question the morality of colonization. As early as 1803, he wrote: *"The right of exterminating, or driving into a corner where they must starve, even the inhabitants of these thinly populated regions, will be questioned in a moral view"* [[Cambridge]]. Yet he held an ambivalent position on slavery: he opposed the slave trade (partly because slave traders used his *Principle of Population* in their defense) but never publicly condemned the institution of slavery itself [[Cambridge]].



## 2. Core Views: "An Essay on the Principle of Population" (1798, 1803)

Malthus's seminal work went through six editions in his lifetime and founded demography as a science.

**Two Postulates:**
1.  *"Food is necessary to the existence of man."*
2.  *"The passion between the sexes is necessary and will remain nearly in its present state."*

**Mathematical Metaphor of Growth:**
| Growth Type | Formula | Example |
|-------------|---------|---------|
| Population | $N_t = N_0 \cdot 2^t$ (geometric progression) | 1, 2, 4, 8, 16... |
| Resources | $F_t = F_0 + k \cdot t$ (arithmetic progression) | 1, 2, 3, 4, 5... |

**Types of Checks:**
*   **Preventive checks:** Moral restraint, late marriage, conscious fertility limitation — *the only ones Malthus deemed morally acceptable*.
*   **Positive checks** (i.e., increasing mortality): War, epidemics, famine, poverty — *natural but tragic correction mechanisms*.

**Important nuance:** Malthus did not "blame the poor." He emphasized that *the wealthy*, by having large families, exacerbated resource scarcity for the poor, and thus bore moral responsibility to limit their own reproduction [[Cambridge]].




## 3. The Exponential Growth Model: Mathematical Formalization

Malthus did not use differential equations in modern notation, but his logic is readily formalized:

$$
\frac{dN}{dt} = r \cdot N, \quad N(0) = N_0
$$

**Solution:**
$$
N(t) = N_0 \cdot e^{rt}, \quad r = b - d
$$

**Where:**
*   $b$ — per-capita birth rate
*   $d$ — per-capita death rate
*   $r > 0$ → unlimited growth; $r < 0$ → extinction

**Model Limitations:**
*   Ignores environmental carrying capacity ($K$)
*   Neglects population age structure
*   Assumes constant $b$ and $d$

**In Julia:** This model is an ideal first case for practicing `DifferentialEquations.jl`, visualization with `Plots.jl`, and parameter sensitivity analysis.



## 4. From Malthusianism to Eugenics: Ideological Evolution

### 4.1. Francis Galton and the Birth of the Term
**Francis Galton** (1822–1911), Charles Darwin's cousin, coined **"eugenics"** in 1883, defining it as *"the study of agencies under social control that may improve or impair the racial qualities of future generations either physically or mentally"* [[Genome.gov]].

Galton, a statistician and demographer, sought to apply quantitative methods to human heredity, relying on (often erroneous) interpretations of Mendelian genetics.



### 4.2. Eugenicist Scientists: Examples and Context
| Scientist | Contribution | Eugenic Stance |
|-----------|-------------|----------------|
| **Karl Pearson** (1857–1936) | Founder of mathematical statistics, correlation, χ²-test | Believed statistics should serve "racial improvement"; headed the Eugenics chair at UCL |
| **Ronald Fisher** (1890–1962) | Founder of modern statistics (ANOVA, maximum likelihood), population genetics | Active eugenics advocate; founded Cambridge Eugenics Society; argued "lower classes" reproduced faster, threatening national quality [[33]][[38]] |
| **Charles Darwin** (1809–1882) | Theory of evolution | Not a political eugenicist, but his "struggle for existence" was appropriated by social Darwinists |
| **Alexis Carrel** (1873–1944) | Nobel laureate in physiology, transplantation pioneer | Supported eugenic measures including euthanasia of the "unfit" in occupied France |



### 4.3. Eugenics vs. Genetics: Fundamental Distinctions
| Criterion | **Historical Eugenics** | **Modern Genetics** |
|-----------|------------------------|---------------------|
| **Scientific Basis** | Simplified, often erroneous understanding of heredity; conflation of social and biological traits | Molecular biology, genomics, epigenetics, statistical genetics |
| **Methodology** | Selection, sterilization, segregation based on phenotype and social status | GWAS, CRISPR, prenatal screening, genetic counseling |
| **Ethical Framework** | Collectivist, state-coercive, hierarchical | Individual autonomy, informed consent, human rights protection |
| **Goal** | "Racial improvement" via exclusion of the "undesirable" | Understanding disease mechanisms, personalized medicine, expanding reproductive choice |

**Important:** Modern genetic counseling and prenatal diagnosis are *sometimes* criticized as "new eugenics," especially regarding selective termination for genetic conditions (e.g., Down syndrome) [[Genome.gov]]. However, the key distinction lies in voluntary decision-making and absence of state coercion.



### 4.4. Eugenics in 20th-Century Ideology: Debates
*   **USA:** Forced sterilization laws for the "unfit" (60,000+ victims by the 1970s), Immigration Act of 1924 restricting "non-Nordic" groups [[Genome.gov]].
*   **Nazi Germany:** Most radical implementation — "T-4" euthanasia program, Holocaust, Nuremberg racial laws. *Crucially:* Nazi racial hygiene used scientific language but was an ideological distortion, not a scientific conclusion.
*   **Post-1945:** Official eugenics was discredited, but elements persisted in family planning policies, demographic programs in developing countries, and "population quality" discourses.

**Contemporary Critique:** Scholars emphasize that eugenics was not "a mistake by a few scientists" but a *broad intellectual movement* supported by statisticians, biologists, economists, and reformers worldwide [[Genome.gov]]. Understanding this context is vital for critically assessing modern biotechnologies.


## 5. Neo-Malthusianism and Modern Ecology: From "Limits to Growth" to Climate Crisis

### 5.1. Revival of Malthusianism in the 20th Century
*   **1968:** Formation of the Club of Rome.
*   **1972:** Publication of **"The Limits to Growth"** (Meadows et al.) — a system dynamics model (World3) predicting industrial civilization collapse by mid-21st century under current trends in population, industrialization, pollution, resource depletion, and food production [[Spash]].

**Key Neo-Malthusian Ideas:**
*   Exponential growth in a finite system leads to crisis.
*   Technological progress does not eliminate but merely *postpones* limits, creating new problems (e.g., Green Revolution → soil erosion, fossil fuel dependency).
*   Solution lies not in growth but in *stabilization* and transition to qualitative, not quantitative, development.



### 5.2. Critique and Counter-Movements
**"Cornucopians"** — optimists believing in boundless human innovative potential:
*   **Julian Simon** (1932–1998): Argued "human capital" is the ultimate resource; population growth stimulates innovation; resource prices fall long-term, not rise.
*   **Argument:** Market signals and technological progress solve scarcity; "limits" are institutional, not physical.

**Ecological Economics** (third position):
*   Acknowledges physical limits but emphasizes *distribution*, *institutional reform*, *value shifts*.
*   Key concepts: sustainable development, planetary boundaries (Rockström et al., 2009), economic dematerialization.



### 5.3. Contemporary Controversial Viewpoints
| Position | Core Arguments | Criticisms |
|----------|---------------|------------|
| **Eco-modernism** | Technological solutions (renewables, CCS, vertical farming) enable growth without planetary harm | Ignores rebound effects, socio-political barriers to adoption |
| **Degrowth** | Conscious rejection of GDP-oriented growth in wealthy nations for ecological justice | Accused of utopianism, neglecting developing countries' needs |
| **Adaptive Pragmatism** | Focus on adapting to inevitable changes (climate, migration), not prevention | Risks capitulation to systemic problems, exacerbating inequality |
| **Techno-optimism (AI, geoengineering)** | AI and geoengineering will solve climate crisis; growth generates solutions | Ethical risks, unpredictable side effects, moral hazard |



### 5.4. Connection to Julia-based Modeling
Modern environmental models (e.g., Integrated Assessment Models, IAMs) employ:
*   Systems of ODEs/DAEs for "climate–economy–population" interlinkages
*   Agent-based modeling for heterogeneous actor behavior
*   Bayesian methods for parameter uncertainty quantification

**Example Challenge:** How to model *social limits to growth* (per F. Hirsch) — where rising welfare encounters positional goods (status, attention, space) that do not scale?


## 6. Conclusion: Why This Matters for Biology and Medicine Modeling

1.  **The Malthusian model** is a foundational building block for more complex models (logistic, Lotka–Volterra, SIR).
2.  **The history of eugenics** warns against *scientific reductionism*: complex social and medical phenomena cannot be reduced to simple genetic determinants.
3.  **Neo-Malthusian debates** teach critical evaluation of model assumptions: What limits do we encode? Who bears the costs? How do we handle uncertainty?
4.  **Julia as a tool** enables rapid prototyping, testing, and visualization of these ideas, making theoretical discussions *computationally concrete*.

> *"A model is not a forecast, but a tool for thinking. Understanding the history of ideas helps ask models the right questions."*

# Racial Classifications in Malthusianism and Eugenics: A Historical-Critical Review
## For Julia-based Mathematical Modeling Course — From the Perspective of Modern Science

> **Important Disclaimer**: This section provides a *historical analysis* of discredited pseudoscientific theories. Modern genetics, anthropology, and demography reject the biological concept of race as a scientific category for explaining social, intellectual, or cultural differences between populations [[24]]. This material is included solely for educational purposes — to understand the history of science, critically analyze models, and reflect ethically on modeling practices.



## 1. Racial Hierarchies in Classical Malthusianism

### 1.1. Malthus and the Imperial Context
Thomas Robert Malthus did not formulate an explicit racial theory in the modern sense, yet his work developed within the context of British imperialism, where demographic arguments were used to justify colonial expansion [[3]].

| Group | Characterization in Malthusian discourse | Source of argumentation |
|-------|------------------------------------------|------------------------|
| **English / Europeans** | "Civilized": capable of moral restraint (preventive checks), rational resource management, delayed marriage | *Essay on Population* (1803), chapters on Europe |
| **"Savage" peoples** (indigenous Americas, Australia, Africa) | Subject primarily to *positive checks*: famine, war, infanticide; incapable of moral self-restraint due to "low stage of civilization" | Travel accounts by Cook, Park, Banks, cited by Malthus [[1]] |
| **Chinese, Indians** | "Semi-civilized": high population density but limited resources; demographic pressure used to explain poverty, not colonial exploitation | Statistical compilations from East India Company reports |

**Key point**: Malthus interpreted differences in fertility and mortality not as biological, but as *cultural-institutional*. However, his framework was easily instrumentalized: if "inferior races" were incapable of moral restraint, their suffering appeared as a "natural" consequence of natural law, not policy [[33]].

### 1.2. Mathematical Metaphor as a Tool of Hierarchization
The exponential growth formula $N(t) = N_0 e^{rt}$ is mathematically neutral. But in Malthusian discourse, the parameter $r$ (growth rate) was implicitly linked to "level of civilization":
*   High $r$ → "uncontrolled reproduction" → sign of "backwardness"
*   Low $r$ → "moral restraint" → sign of "progress"

> **Methodological lesson for Julia**: When modeling demographic processes, it is essential to explicitly separate the *mathematical structure of the model* from the *interpretation of parameters*. The same code can serve both epidemic analysis and justification of discrimination — depending on assumptions and context.

### 2.2. Eugenics vs. Genetics: Fundamental Distinctions

| Criterion | Historical Eugenics | Modern Population Genetics |
|-----------|---------------------|---------------------------|
| **Unit of Analysis** | "Race" as biologically discrete, hierarchical category | Population as continuous gradient of genetic variation |
| **Method** | Phenotypic observation, craniometry, social stratification | GWAS, sequencing, statistical analysis of allele frequencies |
| **Interpretation of Differences** | Genetic determinism: intelligence, morality, culture "encoded" in race | Gene-environment interaction; epigenetics; cultural evolution |
| **Ethical Imperative** | Collective "racial improvement" via state control | Individual autonomy, informed consent, human rights protection |

> **Important**: Modern research shows that >99% of human genetic variation exists *within* any условные "racial" groups, not *between* them. The concept of biological race has no operational meaning in medicine or demography without accounting for social context [[24]].

## 3. Critical Reflection: Why This Matters for Modeling?

### 3.1. Historical Lessons for Model Developers
1.  **Parameters are not neutral**: Assigning values of $r$, $K$, $b$, $d$ to different population groups can reproduce hidden biases if data sources and assumptions are not scrutinized.
2.  **Aggregation is dangerous**: Modeling "Chinese", "Indians", or "Africans" as homogeneous blocks ignores internal diversity and leads to erroneous predictions.
3.  **Interpretation context**: The same exponential growth model can describe either infection spread or "overpopulation threat" — the choice of language and visualization has ethical dimensions.

### 3.2. Practical Recommendations for Julia Code
```julia
# Example: explicit documentation of model assumptions
"""
    malthus_model(N0, r, t; population_group="generic")

Solves dN/dt = r*N with initial condition N(0)=N0.

# Arguments
- `population_group::String`: label for documentation (does not affect calculations!).
  Important: parameter `r` should be estimated from local demographic data,
  not assigned based on ethnic, racial, or national affiliation.

# Ethical warning
Historically, growth parameters were used to justify discriminatory policies.
Always cite data sources and check whether the model reproduces
hidden biases of original studies.
"""
function malthus_model(N0, r, t; population_group="generic")
    return N0 * exp(r * t)
end

## 4. Sources and Further Reading

Scholarly Reviews and Historical Studies

.
Cussen O. Prophet of the Past: Malthus and the Margins of History // London Review of Books, 2024. — Analysis of imperial context of Malthusianism 
www.lrb.co.uk

.
Schneider W. Toward the Improvement of the Human Race: The History of Eugenics in France // Journal of Modern History, 1982. — French eugenics and colonialism 
www.journals.uchicago.edu

.
Stone D. Race in British Eugenics // European History Quarterly, 2001. — British eugenics and racial classifications 
Sage

.
Weindling P. Health, Race and German Politics Between National Unification and Nazism. Cambridge UP, 1989. — German "racial hygiene".
Gould S.J. The Mismeasure of Man. Norton, 1996. — Critique of craniometry and racial hierarchies.
Modern Positions of Scientific Communities
American Association of Physical Anthropologists. Statement on Race & Racism, 2019. — Rejection of biological concept of race.
UNESCO. The Race Question, 1950 (updated 2021). — International consensus on social, not biological, nature of race 
en.wikipedia.org

.
National Human Genome Research Institute. Eugenics and Scientific Racism. — Educational materials on history and ethics [[Genome.gov]].

For Practical Modeling in Julia

Rackauckas C., Nie Q. DifferentialEquations.jl – A Performant and Feature-Rich Ecosystem for Solving Differential Equations in Julia. Journal of Open Research Software, 2017.

Ethical Guidelines for Computational Demography (forthcoming, International Union for the Scientific Study of Population, 2025).

## Concluding Principle:
"A model is not a mirror of reality, but a tool for asking questions. The history of eugenics teaches: the most dangerous models are not those mathematically flawed, but those formally impeccable yet ethically blind."
Material prepared for inclusion in Jupyter/Pluto notebooks for the course "Mathematical Modeling in Biology and Medicine using Julia". Recommended to accompany code with links to ethical guidelines and critical commentary.
