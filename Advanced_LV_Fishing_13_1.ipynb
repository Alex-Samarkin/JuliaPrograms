#### Schaefer Model (1954)

$$ \frac{dN}{dt} = rN\left(1 - \frac{N}{K}\right) - qEN $$

## 2. Environment and Data Preparation

````julia
"""
–°—Ç—Ä—É–∫—Ç—É—Ä–∞ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤–∏–¥–∞ —Ä—ã–±
Structure for storing fish species parameters

–ü–æ–ª—è / Fields:
- name: –ù–∞—É—á–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ / Scientific name
- name_ru: –†—É—Å—Å–∫–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ / Russian name
- r: –°–∫–æ—Ä–æ—Å—Ç—å —Ä–æ—Å—Ç–∞ / Growth rate (year‚Åª¬π)
- K: –Å–º–∫–æ—Å—Ç—å —Å—Ä–µ–¥—ã / Carrying capacity (tons)
- T_opt: –û–ø—Ç–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ / Optimal temperature (¬∞C)
- T_min: –ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ / Min temperature (¬∞C)
- T_max: –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ / Max temperature (¬∞C)
- initial_biomass: –ù–∞—á–∞–ª—å–Ω–∞—è –±–∏–æ–º–∞—Å—Å–∞ / Initial biomass (tons)
- catch_quota: –ö–≤–æ—Ç–∞ –≤—ã–ª–æ–≤–∞ / Catch quota (tons/year)
- trophic_level: –¢—Ä–æ—Ñ–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å / Trophic level
"""
struct FishSpecies
    name::String
    name_ru::String
    r::Float64
    K::Float64
    T_opt::Float64
    T_min::Float64
    T_max::Float64
    initial_biomass::Float64
    catch_quota::Float64
    trophic_level::Float64
end

println("‚úÖ –°—Ç—Ä—É–∫—Ç—É—Ä–∞ FishSpecies –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ / FishSpecies structure defined")
````

````julia
# ==============================================================================
# –°–û–ó–î–ê–ù–ò–ï –û–ë–™–ï–ö–¢–û–í –í–ò–î–û–í –†–´–ë
# CREATING FISH SPECIES OBJECTS
# ==============================================================================

# –ê—Ç–ª–∞–Ω—Ç–∏—á–µ—Å–∫–∞—è —Ç—Ä–µ—Å–∫–∞ / Atlantic Cod
species_cod = FishSpecies(
    "Gadus morhua",           # –ù–∞—É—á–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ
    "–ê—Ç–ª–∞–Ω—Ç–∏—á–µ—Å–∫–∞—è —Ç—Ä–µ—Å–∫–∞",   # –†—É—Å—Å–∫–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ
    0.4,                      # r (–≥–æ–¥‚Åª¬π)
    3_000_000,                # K (—Ç–æ–Ω–Ω)
    6.0,                      # T_opt (¬∞C)
    2.0,                      # T_min (¬∞C)
    12.0,                     # T_max (¬∞C)
    1_500_000,                # –ù–∞—á–∞–ª—å–Ω–∞—è –±–∏–æ–º–∞—Å—Å–∞ (—Ç–æ–Ω–Ω)
    150_000,                  # –ö–≤–æ—Ç–∞ (—Ç–æ–Ω–Ω/–≥–æ–¥)
    4.5                       # –¢—Ä–æ—Ñ–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å
)

# –ê–Ω—á–æ—É—Å / Anchovy
species_anchovy = FishSpecies(
    "Engraulis encrasicolus",
    "–ê–Ω—á–æ—É—Å",
    1.0,
    8_000_000,
    17.0,
    10.0,
    25.0,
    4_000_000,
    400_000,
    3.0
)

# –ú–æ–π–≤–∞ / Capelin
species_capelin = FishSpecies(
    "Mallotus villosus",
    "–ú–æ–π–≤–∞",
    0.75,
    5_000_000,
    4.0,
    0.0,
    10.0,
    2_500_000,
    250_000,
    3.5
)

# –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –≤–∏–¥–æ–≤
species_list = [species_cod, species_anchovy, species_capelin]

println("üêü –ó–∞–≥—Ä—É–∂–µ–Ω–æ –≤–∏–¥–æ–≤: ", length(species_list))
println()
for (i, sp) in enumerate(species_list)
    println("   $(i). $(sp.name_ru) ($(sp.name))")
    println("      ‚Ä¢ –°–∫–æ—Ä–æ—Å—Ç—å —Ä–æ—Å—Ç–∞ r = $(sp.r) –≥–æ–¥‚Åª¬π")
    println("      ‚Ä¢ –Å–º–∫–æ—Å—Ç—å K = $(sp.K/1_000_000) –º–ª–Ω —Ç–æ–Ω–Ω")
    println("      ‚Ä¢ T_opt = $(sp.T_opt)¬∞C [$(sp.T_min)¬∞C - $(sp.T_max)¬∞C]")
    println()
end
````

````julia
# ==============================================================================
# –ì–†–ê–§–ò–ö 2: –¢–ï–ú–ü–ï–†–ê–¢–£–†–ù–´–ï –ù–ò–®–ò –í–ò–î–û–í
# PLOT 2: SPECIES TEMPERATURE NICHES
# ==============================================================================

species_colors = [:blue, :green, :orange]

p_temp_niche = plot(title = "–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω—ã–µ –Ω–∏—à–∏ –≤–∏–¥–æ–≤\nSpecies Temperature Niches",
                    titlefont = (14, :bold),
                    xlabel = "–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –≤–æ–¥—ã (¬∞C) / Water Temperature (¬∞C)",
                    ylabel = "–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –ø—Ä–∏–≥–æ–¥–Ω–æ—Å—Ç—å / Relative Suitability",
                    size = (1000, 600),
                    legend = :topright)

# –§—É–Ω–∫—Ü–∏—è –ì–∞—É—Å—Å–∞ –¥–ª—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω–æ–π –Ω–∏—à–∏
function temp_niche(T, T_opt, T_min, T_max)
    œÉ = (T_max - T_min) / 4
    return exp(-((T - T_opt)^2) / (2 * œÉ^2))
end

T_range = -2:0.1:28

for (i, sp) in enumerate(species_list)
    niche_curve = [temp_niche(T, sp.T_opt, sp.T_min, sp.T_max) for T in T_range]
    plot!(p_temp_niche, T_range, niche_curve,
          label = sp.name_ru,
          color = species_colors[i],
          linewidth = 3,
          alpha = 0.7)
    
    # –í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–∞—è –ª–∏–Ω–∏—è –æ–ø—Ç–∏–º–∞–ª—å–Ω–æ–π —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
    vline!(p_temp_niche, [sp.T_opt], 
           color = species_colors[i], 
           linestyle = :dash, 
           label = "")
end

# –ê–Ω–Ω–æ—Ç–∞—Ü–∏–∏
annotate!(p_temp_niche, 
          [(species_cod.T_opt, 0.9, text("–¢—Ä–µ—Å–∫–∞ optimum", 9, :blue)),
           (species_anchovy.T_opt, 0.9, text("–ê–Ω—á–æ—É—Å optimum", 9, :green)),
           (species_capelin.T_opt, 0.9, text("–ú–æ–π–≤–∞ optimum", 9, :orange))])

display(p_temp_niche)
````

### 2.2. –ú–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –≤–æ–¥—ã

### 2.2. Water Temperature Modeling

````julia
# ==============================================================================
# –§–£–ù–ö–¶–ò–ò –ú–û–î–ï–õ–ò–†–û–í–ê–ù–ò–Ø –¢–ï–ú–ü–ï–†–ê–¢–£–†–´
# TEMPERATURE MODELING FUNCTIONS
# ==============================================================================

"""
–§—É–Ω–∫—Ü–∏—è –¥–ª—è –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –≤–æ–¥—ã —Å —É—á—ë—Ç–æ–º –∫–ª–∏–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ç—Ä–µ–Ω–¥–∞
Function for modeling water temperature with climate trend

–ê—Ä–≥—É–º–µ–Ω—Ç—ã / Arguments:
- t: –≤—Ä–µ–º—è (–ª–µ—Ç) / time (years)
- base_temp: –±–∞–∑–æ–≤–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ (¬∞C) / base temperature
- trend: —Ç—Ä–µ–Ω–¥ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è (¬∞C/–≥–æ–¥) / warming trend
- seasonal_amp: –∞–º–ø–ª–∏—Ç—É–¥–∞ —Å–µ–∑–æ–Ω–Ω—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π / seasonal amplitude

–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç: —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –≤–æ–¥—ã (¬∞C)
Returns: water temperature (¬∞C)
"""
function water_temperature(t::Real; 
                           base_temp::Float64 = 8.0,
                           trend::Float64 = 0.03,
                           seasonal_amp::Float64 = 3.0)
    # –ü—Ä–∏–≤–æ–¥–∏–º –∫ –ø–ª–∞–≤–∞—é—â–µ–π —Ç–æ—á–∫–µ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏–π
    t = float(t)
    # –ë–∞–∑–æ–≤–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ + —Ç—Ä–µ–Ω–¥ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è + —Å–µ–∑–æ–Ω–Ω—ã–µ –∫–æ–ª–µ–±–∞–Ω–∏—è
    temp = base_temp + trend * t + seasonal_amp * sin(2œÄ * t / 12)
    return temp
end
````

````julia
# ==============================================================================
# –ì–†–ê–§–ò–ö 3: –ú–û–î–ï–õ–¨ –¢–ï–ú–ü–ï–†–ê–¢–£–†–´ –í–û –í–†–ï–ú–ï–ù–ò
# PLOT 3: WATER TEMPERATURE MODEL OVER TIME
# ==============================================================================

test_times = 0:0.1:50
test_temps_trend = [water_temperature(t, base_temp=8.0, trend=0.03, seasonal_amp=0.0) for t in test_times]
test_temps_seasonal = [water_temperature(t, base_temp=8.0, trend=0.03, seasonal_amp=3.0) for t in test_times]

p_temp_model = plot(title = "–ú–æ–¥–µ–ª—å –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –≤–æ–¥—ã\nWater Temperature Change Model",
                    titlefont = (14, :bold),
                    xlabel = "–í—Ä–µ–º—è (–≥–æ–¥—ã) / Time (years)",
                    ylabel = "–¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ (¬∞C) / Temperature (¬∞C)",
                    size = (1000, 600),
                    legend = :topleft)

# –õ–∏–Ω–∏—è —Ç—Ä–µ–Ω–¥–∞
plot!(p_temp_model, test_times, test_temps_trend,
      label = "–¢—Ä–µ–Ω–¥ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è / Warming trend",
      color = :red,
      linewidth = 3,
      linestyle = :dash)

# –ü–æ–ª–Ω–∞—è –º–æ–¥–µ–ª—å —Å —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å—é
plot!(p_temp_model, test_times, test_temps_seasonal,
      label = "–° —Å–µ–∑–æ–Ω–Ω—ã–º–∏ –∫–æ–ª–µ–±–∞–Ω–∏—è–º–∏ / With seasonal fluctuations",
      color = :blue,
      linewidth = 2,
      alpha = 0.6)

# –ì–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏ –¥–ª—è –æ–ø—Ç–∏–º—É–º–æ–≤
hline!(p_temp_model, [species_cod.T_opt], 
       label = "–û–ø—Ç–∏–º—É–º —Ç—Ä–µ—Å–∫–∏ ($(species_cod.T_opt)¬∞C)", 
       linestyle = :dash, 
       color = :blue,
       linewidth = 2)
hline!(p_temp_model, [species_anchovy.T_opt], 
       label = "–û–ø—Ç–∏–º—É–º –∞–Ω—á–æ—É—Å–∞ ($(species_anchovy.T_opt)¬∞C)", 
       linestyle = :dash, 
       color = :green,
       linewidth = 2)
hline!(p_temp_model, [species_capelin.T_opt], 
       label = "–û–ø—Ç–∏–º—É–º –º–æ–π–≤—ã ($(species_capelin.T_opt)¬∞C)", 
       linestyle = :dash, 
       color = :orange,
       linewidth = 2)

# –ê–Ω–Ω–æ—Ç–∞—Ü–∏—è —Ç—Ä–µ–Ω–¥–∞
annotate!(p_temp_model, 40, 10, text("–ü–æ—Ç–µ–ø–ª–µ–Ω–∏–µ +0.03¬∞C/–≥–æ–¥\nWarming +0.03¬∞C/year", 10, :red))

display(p_temp_model)

println("üìà –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞: $(round(test_temps_seasonal[1], digits=2))¬∞C")
println("üìà –¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ —á–µ—Ä–µ–∑ 50 –ª–µ—Ç: $(round(test_temps_seasonal[end], digits=2))¬∞C")
println("üìà –û–±—â–µ–µ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏–µ: $(round(test_temps_seasonal[end] - test_temps_seasonal[1], digits=2))¬∞C")
````

## 3. –û–¥–Ω–æ-–≤–∏–¥–æ–≤–∞—è –º–æ–¥–µ–ª—å —Å –∫–ª–∏–º–∞—Ç–æ–º

## 3. Single-Species Model with Climate

### 3.1. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏

### 3.1. Mathematical Problem Statement

–ú–æ–¥–µ–ª—å —Ä–æ—Å—Ç–∞ –ø–æ–ø—É–ª—è—Ü–∏–∏ —Å –∫–ª–∏–º–∞—Ç–∏—á–µ—Å–∫–∏–º –≤–ª–∏—è–Ω–∏–µ–º:

Population growth model with climate impact:

$$ \frac{dN}{dt} = r \cdot f(T) \cdot N \cdot \left(1 - \frac{N}{K \cdot f(T)}\right) - H $$

–ì–¥–µ:

- $f(T)$ ‚Äî —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤–ª–∏—è–Ω–∏—è / temperature impact function
- $K(T) = K_0 \cdot f(T)$ ‚Äî —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω–æ-–∑–∞–≤–∏—Å–∏–º–∞—è —ë–º–∫–æ—Å—Ç—å —Å—Ä–µ–¥—ã
- $H$ ‚Äî –ø–æ—Å—Ç–æ—è–Ω–Ω—ã–π –≤—ã–ª–æ–≤ (–∫–≤–æ—Ç–∞)

````julia
# ==============================================================================
# –û–î–ù–û–í–ò–î–û–í–ê–Ø –ú–û–î–ï–õ–¨ –° –ö–õ–ò–ú–ê–¢–ò–ß–ï–°–ö–ò–ú –í–õ–ò–Ø–ù–ò–ï–ú
# SINGLE-SPECIES MODEL WITH CLIMATE IMPACT
# ==============================================================================

"""
–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ –û–î–£ –¥–ª—è –æ–¥–Ω–æ–≤–∏–¥–æ–≤–æ–π –º–æ–¥–µ–ª–∏
ODE right-hand side function for single-species model

–ê—Ä–≥—É–º–µ–Ω—Ç—ã / Arguments:
- du: –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è / derivative
- u: –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ —Å–æ—Å—Ç–æ—è–Ω–∏—è / state variables (–±–∏–æ–º–∞—Å—Å–∞ N)
- p: –ø–∞—Ä–∞–º–µ—Ç—Ä—ã / parameters
- t: –≤—Ä–µ–º—è / time

–ü–∞—Ä–∞–º–µ—Ç—Ä—ã / Parameters:
p[1] = r: —Å–∫–æ—Ä–æ—Å—Ç—å —Ä–æ—Å—Ç–∞
p[2] = K0: –±–∞–∑–æ–≤–∞—è —ë–º–∫–æ—Å—Ç—å
p[3] = H: –∫–≤–æ—Ç–∞ –≤—ã–ª–æ–≤–∞
p[4] = temp_base: –±–∞–∑–æ–≤–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
p[5] = temp_trend: —Ç—Ä–µ–Ω–¥ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è
p[6] = T_opt: –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
p[7] = T_min: –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
p[8] = T_max: –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
"""
function single_species_climate!(du, u, p, t)
    N = u[1]
    r, K0, H, temp_base, temp_trend, T_opt, T_min, T_max = p
    
    # –¢–µ–∫—É—â–∞—è —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞
    T = water_temperature(t, base_temp = temp_base, trend = temp_trend)
    
    # –í–ª–∏—è–Ω–∏–µ —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã –Ω–∞ —Ä–æ—Å—Ç
    temp_eff = temperature_effect(T, T_opt, T_min, T_max)
    
    # –Å–º–∫–æ—Å—Ç—å —Å—Ä–µ–¥—ã –∑–∞–≤–∏—Å–∏—Ç –æ—Ç —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä—ã
    K_t = K0 * temp_eff
    
    # –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∏–π —Ä–æ—Å—Ç —Å –≤—ã–ª–æ–≤–æ–º
    if N > 0.1
        growth = r * temp_eff * N * (1.0 - N / max(K_t, 1000))
        du[1] = growth - H
    else
        du[1] = 0.0
    end
    
    # –ó–∞—â–∏—Ç–∞ –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–π –±–∏–æ–º–∞—Å—Å—ã
    if du[1] < 0 && N < 1000
        du[1] = max(du[1], -N/10)
    end
end

println("‚úÖ –§—É–Ω–∫—Ü–∏—è single_species_climate! –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞")
````

````julia
# ==============================================================================
# –ü–ê–†–ê–ú–ï–¢–†–´ –ò –ù–ê–ß–ê–õ–¨–ù–´–ï –£–°–õ–û–í–ò–Ø –î–õ–Ø –¢–†–ï–°–ö–ò
# PARAMETERS AND INITIAL CONDITIONS FOR COD
# ==============================================================================

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Ç—Ä–µ—Å–∫–∏
p_cod = [
    species_cod.r,              # r
    species_cod.K,              # K0
    species_cod.catch_quota,    # H
    6.0,                        # temp_base
    0.03,                       # temp_trend (–ø–æ—Ç–µ–ø–ª–µ–Ω–∏–µ)
    species_cod.T_opt,          # T_opt
    species_cod.T_min,          # T_min
    species_cod.T_max           # T_max
]

# –ù–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è –∏ –≤—Ä–µ–º—è
u0_cod = [species_cod.initial_biomass]
tspan = (0.0, 50.0)

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏
prob_cod = ODEProblem(single_species_climate!, u0_cod, tspan, p_cod)
sol_cod = solve(prob_cod, Tsit5(), reltol = 1e-8, abstol = 1e-8, saveat = 0.5)

println("‚úÖ –†–µ—à–µ–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–æ / Solution obtained")
println("   ‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫: ", length(sol_cod.t))
println("   ‚Ä¢ –ù–∞—á–∞–ª—å–Ω–∞—è –±–∏–æ–º–∞—Å—Å–∞: $(u0_cod[1] / 1_000_000) –º–ª–Ω —Ç–æ–Ω–Ω")
println("   ‚Ä¢ –§–∏–Ω–∞–ª—å–Ω–∞—è –±–∏–æ–º–∞—Å—Å–∞: $(sol_cod[1,end] / 1_000_000) –º–ª–Ω —Ç–æ–Ω–Ω")
````

### 3.3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

## 4. –ú–Ω–æ–≥–æ–≤–∏–¥–æ–≤–∞—è –º–æ–¥–µ–ª—å (–≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–µ –≤–∏–¥–æ–≤)

## 4. Multi-Species Model (Species Interaction)

### 4.1. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –º–æ–¥–µ–ª—å

### 4.1. Mathematical Model

–ú–Ω–æ–≥–æ–≤–∏–¥–æ–≤–∞—è –º–æ–¥–µ–ª—å —Å –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–µ–º –∏ –∫–ª–∏–º–∞—Ç–æ–º:

Multi-species model with interaction and climate:

$$ \frac{dN_i}{dt} = r_i \cdot f_i(T) \cdot N_i \cdot \left(1 - \sum_{j=1}^{n} \frac{\alpha_{ij} N_j}{K_i}\right) - H_i $$

**–ú–∞—Ç—Ä–∏—Ü–∞ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–π / Interaction Matrix:**

$$ \alpha = \begin{pmatrix} 
1.0 & 0.1 & 0.3 \\
0.05 & 1.0 & 0.2 \\
0.2 & 0.1 & 1.0 
\end{pmatrix} $$

–ì–¥–µ $\alpha_{ij}$ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–ª–∏—è–Ω–∏–µ –≤–∏–¥–∞ $j$ –Ω–∞ –≤–∏–¥ $i$.

````julia
# ==============================================================================
# –ú–ù–û–ì–û–í–ò–î–û–í–ê–Ø –ú–û–î–ï–õ–¨ –õ–û–¢–ö–ò-–í–û–õ–¨–¢–ï–†–†–´ –° –ö–õ–ò–ú–ê–¢–û–ú
# MULTI-SPECIES LOTKA-VOLTERRA MODEL WITH CLIMATE
# ==============================================================================

"""
–§—É–Ω–∫—Ü–∏—è –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ –û–î–£ –¥–ª—è –º–Ω–æ–≥–æ–≤–∏–¥–æ–≤–æ–π –º–æ–¥–µ–ª–∏
ODE right-hand side function for multi-species model

–ê—Ä–≥—É–º–µ–Ω—Ç—ã / Arguments:
- du: –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ / derivatives (–≤–µ–∫—Ç–æ—Ä –¥–ª–∏–Ω—ã n_species)
- u: –±–∏–æ–º–∞—Å—Å—ã –≤–∏–¥–æ–≤ / species biomasses (–≤–µ–∫—Ç–æ—Ä –¥–ª–∏–Ω—ã n_species)
- p: –ø–∞—Ä–∞–º–µ—Ç—Ä—ã / parameters
- t: –≤—Ä–µ–º—è / time

–°—Ç—Ä—É–∫—Ç—É—Ä–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ / Parameter structure:
p[1:n] = r: —Å–∫–æ—Ä–æ—Å—Ç–∏ —Ä–æ—Å—Ç–∞
p[n+1:2n] = K: —ë–º–∫–æ—Å—Ç–∏ —Å—Ä–µ–¥—ã
p[2n+1:3n] = H: –∫–≤–æ—Ç—ã –≤—ã–ª–æ–≤–∞
p[3n+1:end] = —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã (base, trend, T_opt –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞)
"""
function multi_species_climate!(du, u, p, t)
    n_species = length(u)
    
    # –†–∞—Å–ø–∞–∫–æ–≤–∫–∞ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    r = p[1:n_species]
    K = p[n_species+1:2*n_species]
    H = p[2*n_species+1:3*n_species]
    temp_params = p[3*n_species+1:end]
    
    # –ú–∞—Ç—Ä–∏—Ü–∞ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏—è –≤–∏–¥–æ–≤
    # Œ±·µ¢: –≤–ª–∏—è–Ω–∏–µ –≤–∏–¥–∞ j –Ω–∞ –≤–∏–¥ i
    Œ± = [
        1.0   0.1   0.3;    # –¢—Ä–µ—Å–∫–∞: —Å–∏–ª—å–Ω–æ–µ –≤–ª–∏—è–Ω–∏–µ –Ω–∞ –º–æ–π–≤—É (—Ö–∏—â–Ω–∏–∫)
        0.05  1.0   0.2;    # –ê–Ω—á–æ—É—Å: —Å–ª–∞–±–∞—è –∫–æ–Ω–∫—É—Ä–µ–Ω—Ü–∏—è
        0.2   0.1   1.0     # –ú–æ–π–≤–∞: –≤–ª–∏—è–µ—Ç –Ω–∞ —Ç—Ä–µ—Å–∫—É (–∫–æ—Ä–º)
    ]
    
    for i in 1:n_species
        N_i = u[i]
        
        # –¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –¥–ª—è –≤–∏–¥–∞ i
        temp_base = temp_params[3*i-2]
        temp_trend = temp_params[3*i-1]
        T_opt = temp_params[3*i]
        T_min = temp_base - 4
        T_max = temp_base + 6
        
        T = water_temperature(t, base_temp = temp_base, trend = temp_trend)
        temp_eff = temperature_effect(T, T_opt, T_min, T_max)
        
        # –°—É–º–º–∞ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–π
        interaction_sum = 0.0
        for j in 1:n_species
            interaction_sum += Œ±[i,j] * u[j] / K[i]
        end
        
        # –†–æ—Å—Ç —Å —É—á—ë—Ç–æ–º –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏–π –∏ –∫–ª–∏–º–∞—Ç–∞
        if N_i > 1000
            growth = r[i] * temp_eff * N_i * (1.0 - interaction_sum)
            du[i] = growth - H[i]
        else
            du[i] = 0.0
        end
        
        # –ó–∞—â–∏—Ç–∞ –æ—Ç –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
        if du[i] < 0 && N_i < 10000
            du[i] = max(du[i], -N_i/20)
        end
    end
end

println("‚úÖ –§—É–Ω–∫—Ü–∏—è multi_species_climate! –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞")
````

````julia
# ==============================================================================
# –ü–ê–†–ê–ú–ï–¢–†–´ –ú–ù–û–ì–û–í–ò–î–û–í–û–ô –ú–û–î–ï–õ–ò
# MULTI-SPECIES MODEL PARAMETERS
# ======================================================================

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Ç—Ä—ë—Ö –≤–∏–¥–æ–≤
r_params = [species_cod.r, species_anchovy.r, species_capelin.r]
K_params = [species_cod.K, species_anchovy.K, species_capelin.K]

# –û–ü–¢–ò–ú–ò–ó–ò–†–û–í–ê–ù–ù–´–ï –ö–í–û–¢–´ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏ —Å–º–µ—à–∞–Ω–Ω—ã—Ö —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤
# –ö–≤–æ—Ç—ã –ø–æ–¥–æ–≥–Ω–∞–Ω—ã —Ç–∞–∫, —á—Ç–æ–±—ã –ø–æ–∫–∞–∑–∞—Ç—å –∫–∞–∫ —É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –≤–ª–∏—è–µ—Ç –Ω–∞ —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å
# Optimized quotas to demonstrate mixed sustainability scenarios
H_params = [
    species_cod.catch_quota * 0.35,       # –¢—Ä–µ—Å–∫–∞: 35% (52.5k —Ç–æ–Ω–Ω/–≥–æ–¥)
    species_anchovy.catch_quota * 0.45,   # –ê–Ω—á–æ—É—Å: 45% (180k —Ç–æ–Ω–Ω/–≥–æ–¥)
    species_capelin.catch_quota * 0.35    # –ú–æ–π–≤–∞: 35% (87.5k —Ç–æ–Ω–Ω/–≥–æ–¥)
]

# –¢–µ–º–ø–µ—Ä–∞—Ç—É—Ä–Ω—ã–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –≤–∏–¥–∞ (–≤ –≤–∏–¥–µ –æ–¥–Ω–æ–º–µ—Ä–Ω–æ–≥–æ –≤–µ–∫—Ç–æ—Ä–∞)
temp_params = [
    6.0, 0.03, species_cod.T_opt,      # –¢—Ä–µ—Å–∫–∞
    17.0, 0.03, species_anchovy.T_opt, # –ê–Ω—á–æ—É—Å
    4.0, 0.03, species_capelin.T_opt   # –ú–æ–π–≤–∞
]

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –≤—Å–µ—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
p_multi = vcat(r_params, K_params, H_params, temp_params)

# –ù–∞—á–∞–ª—å–Ω—ã–µ –±–∏–æ–º–∞—Å—Å—ã
u0_multi = [species_cod.initial_biomass, 
            species_anchovy.initial_biomass, 
            species_capelin.initial_biomass]

# –°–æ–∑–¥–∞–Ω–∏–µ –∏ —Ä–µ—à–µ–Ω–∏–µ –∑–∞–¥–∞—á–∏
prob_multi = ODEProblem(multi_species_climate!, u0_multi, tspan, p_multi)
sol_multi = solve(prob_multi, Tsit5(), reltol = 1e-8, abstol = 1e-8, saveat = 0.5)

println("‚úÖ –ú–Ω–æ–≥–æ–≤–∏–¥–æ–≤–∞—è –º–æ–¥–µ–ª—å —Ä–µ—à–µ–Ω–∞ / Multi-species model solved")
println("   ‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–¥–æ–≤: 3")
println("   ‚Ä¢ –í—Ä–µ–º–µ–Ω–Ω–æ–π –∏–Ω—Ç–µ—Ä–≤–∞–ª: $(tspan[1]) - $(tspan[2]) –ª–µ—Ç")
println("   ‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫: ", length(sol_multi.t))
````

## 5. –ê–Ω–∞–ª–∏–∑ —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤ —É–ø—Ä–∞–≤–ª–µ–Ω–∏—è

## 5. Management Scenario Analysis


### 5.1. –ü–æ—Å—Ç–∞–Ω–æ–≤–∫–∞ –∑–∞–¥–∞—á–∏ –∞–Ω–∞–ª–∏–∑–∞ —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤

### 5.1. Scenario Analysis Problem Statement

–¶–µ–ª—å: –ò—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç—å –ø–æ–ø—É–ª—è—Ü–∏–π –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö:

- –£—Ä–æ–≤–Ω—è—Ö –∫–≤–æ—Ç –≤—ã–ª–æ–≤–∞ (0.5x - 1.5x –æ—Ç –±–∞–∑–æ–≤–æ–≥–æ)
- –°–∫–æ—Ä–æ—Å—Ç—è—Ö –∫–ª–∏–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è (0 - 0.06¬∞C/–≥–æ–¥)

–ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏: –í—Å–µ –≤–∏–¥—ã —Å–æ—Ö—Ä–∞–Ω—è—é—Ç –±–æ–ª–µ–µ 10% –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–π –±–∏–æ–º–∞—Å—Å—ã —á–µ—Ä–µ–∑ 50 –ª–µ—Ç.

````julia
# ==============================================================================
# –§–£–ù–ö–¶–ò–Ø –ó–ê–ü–£–°–ö–ê –°–¶–ï–ù–ê–†–ò–Ø
# SCENARIO RUN FUNCTION
# ==============================================================================

"""
–§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–ø—É—Å–∫–∞ —Å—Ü–µ–Ω–∞—Ä–∏—è –∏ –≤–æ–∑–≤—Ä–∞—Ç–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
Function to run scenario and return results

–ê—Ä–≥—É–º–µ–Ω—Ç—ã:
- quota_multiplier: –º–Ω–æ–∂–∏—Ç–µ–ª—å –∫–≤–æ—Ç—ã (0.5 - 1.5)
- climate_trend: —Ç—Ä–µ–Ω–¥ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏—è (¬∞C/–≥–æ–¥)

–í–æ–∑–≤—Ä–∞—â–∞–µ—Ç: —Ä–µ—à–µ–Ω–∏–µ –û–î–£
"""
function run_scenario(quota_multiplier::Float64, climate_trend::Float64)
    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–µ –±–∞–∑–æ–≤—ã–µ –∫–≤–æ—Ç—ã –∏–∑ H_params
    _H = H_params .* quota_multiplier
    
    _p = vcat(r_params, K_params, _H, 
              [6.0, climate_trend, species_cod.T_opt,
               17.0, climate_trend, species_anchovy.T_opt,
               4.0, climate_trend, species_capelin.T_opt])
    
    _prob = ODEProblem(multi_species_climate!, u0_multi, tspan, _p)
    _sol = solve(_prob, Tsit5(), reltol = 1e-6, abstol = 1e-6, saveat = 1.0)
    
    return _sol
end

println("‚úÖ –§—É–Ω–∫—Ü–∏—è run_scenario –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞")
````

````julia
# ==============================================================================
# –ú–ê–¢–†–ò–¶–ê –°–¶–ï–ù–ê–†–ò–ï–í
# SCENARIO MATRIX
# ==============================================================================

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤
quota_multipliers = [0.5, 0.75, 1.0, 1.25, 1.5]
climate_trends = [0.0, 0.02, 0.04, 0.06]

# DataFrame –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
scenario_results = DataFrame(
    Quota_Multiplier = Float64[],
    Climate_Trend = Float64[],
    Cod_Final = Float64[],
    Anchovy_Final = Float64[],
    Capelin_Final = Float64[],
    Cod_Min = Float64[],
    Sustainable = Bool[]
)

# –ó–∞–ø—É—Å–∫ –≤—Å–µ—Ö —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤
println("üöÄ –ó–∞–ø—É—Å–∫ –∞–Ω–∞–ª–∏–∑–∞ —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤...")
println("   –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤: $(length(quota_multipliers) * length(climate_trends))")
println()

for qm in quota_multipliers
    for ct in climate_trends
        sol = run_scenario(qm, ct)
        
        cod_final = sol[1,end] / 1_000_000
        anchovy_final = sol[2,end] / 1_000_000
        capelin_final = sol[3,end] / 1_000_000
        cod_min = minimum(sol[1,:]) / 1_000_000
        
        # –ö—Ä–∏—Ç–µ—Ä–∏–π —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏: –≤—Å–µ –≤–∏–¥—ã –≤—ã—à–µ 10% –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–π
        sustainable = (cod_final > 0.1 * u0_multi[1]/1_000_000) &&
                      (anchovy_final > 0.1 * u0_multi[2]/1_000_000) &&
                      (capelin_final > 0.1 * u0_multi[3]/1_000_000)
        
        push!(scenario_results, (qm, ct, cod_final, anchovy_final, 
                                 capelin_final, cod_min, sustainable))
    end
end

println("‚úÖ –ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à–µ–Ω / Analysis complete")
println("   –£—Å—Ç–æ–π—á–∏–≤—ã—Ö —Å—Ü–µ–Ω–∞—Ä–∏–µ–≤: $(sum(scenario_results.Sustainable)) –∏–∑ $(nrow(scenario_results))")
println()
display(scenario_results)
````

````julia
# ==============================================================================
# –î–ò–ê–ì–ù–û–°–¢–ò–ö–ê: –ü–†–û–í–ï–†–ö–ê –ö–†–ò–¢–ï–†–ò–ï–í –£–°–¢–û–ô–ß–ò–í–û–°–¢–ò
# DIAGNOSTICS: SUSTAINABILITY CRITERIA CHECK
# ==============================================================================

println("\n" * "="^70)
println("üîç –î–ò–ê–ì–ù–û–°–¢–ò–ö–ê –ö–†–ò–¢–ï–†–ò–ï–í –£–°–¢–û–ô–ß–ò–í–û–°–¢–ò")
println("="^70)
println()

# –ù–∞—á–∞–ª—å–Ω—ã–µ –ø–æ—Ä–æ–≥–∏ (10% –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–π –±–∏–æ–º–∞—Å—Å—ã)
thresholds = u0_multi .* 0.1
println("–ü–æ—Ä–æ–≥–∏ —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏ (10% –æ—Ç –Ω–∞—á–∞–ª—å–Ω–æ–≥–æ):")
for i in 1:3
    println("   $i. $(species_names[i]): $(thresholds[i]/1_000_000) –º–ª–Ω —Ç–æ–Ω–Ω")
end
println()

# –ü—Ä–æ–≤–µ—Ä–∏–º —Å–∞–º—ã–π –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã–π —Å—Ü–µ–Ω–∞—Ä–∏–π: –º–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –∫–≤–æ—Ç–∞, –Ω—É–ª–µ–≤–æ–µ –ø–æ—Ç–µ–ø–ª–µ–Ω–∏–µ
sol_best = run_scenario(0.5, 0.0)

println("–°–∞–º—ã–π –±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã–π —Å—Ü–µ–Ω–∞—Ä–∏–π (–∫–≤–æ—Ç–∞=0.5x, –ø–æ—Ç–µ–ø–ª–µ–Ω–∏–µ=0¬∞C/–≥–æ–¥):")
for i in 1:3
    final_biomass = sol_best[i,end]
    threshold = thresholds[i]
    percent = (final_biomass / threshold) * 100
    is_sustainable = final_biomass > threshold
    status = is_sustainable ? "‚úÖ –î–ê" : "‚ùå –ù–ï–¢"
    println("   $i. $(species_names[i]):")
    println("      –§–∏–Ω–∞–ª—å–Ω–∞—è –±–∏–æ–º–∞—Å—Å–∞: $(round(final_biomass/1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
    println("      –¢—Ä–µ–±—É–µ—Ç—Å—è: $(round(threshold/1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
    println("      –ü—Ä–æ—Ü–µ–Ω—Ç –æ—Ç –ø–æ—Ä–æ–≥–∞: $(round(percent, digits=1))%")
    println("      –£—Å—Ç–æ–π—á–∏–≤–∞: $status")
end
println()

# –ü—Ä–æ–≤–µ—Ä–∏–º —Ç–∞–∫–∂–µ –≤—Ä–µ–º–µ–Ω–Ω—ã–π —Ä—è–¥ –∑–∞ 50 –ª–µ—Ç
println("–ü–æ–ª–Ω—ã–π –¥–∏–∞–ø–∞–∑–æ–Ω –±–∏–æ–º–∞—Å—Å –∑–∞ 50 –ª–µ—Ç (–±–ª–∞–≥–æ–ø—Ä–∏—è—Ç–Ω—ã–π —Å—Ü–µ–Ω–∞—Ä–∏–π):")
for i in 1:3
    min_b = minimum(sol_best[i,:])
    max_b = maximum(sol_best[i,:])
    println("   $i. $(species_names[i]): –º–∏–Ω=$(round(min_b/1_000_000, digits=3)) –º–ª–Ω, –º–∞–∫—Å=$(round(max_b/1_000_000, digits=3)) –º–ª–Ω")
end

println()
println("="^70)
````

````julia
# ==============================================================================
# –î–ò–ê–ì–ù–û–°–¢–ò–ö–ê 2: –ê–ù–ê–õ–ò–ó –ö–û–ù–ö–£–†–ï–ù–¶–ò–ò –ò –í–ó–ê–ò–ú–û–î–ï–ô–°–¢–í–ò–ô
# DIAGNOSTICS 2: COMPETITION AND INTERACTION ANALYSIS
# ==============================================================================

println("\n" * "="^70)
println("üîç –ê–ù–ê–õ–ò–ó –ú–ê–¢–†–ò–¶–´ –ö–û–ù–ö–£–†–ï–ù–¶–ò–ò –ò –í–ó–ê–ò–ú–û–î–ï–ô–°–¢–í–ò–Ø")
println("="^70)
println()

# –ú–∞—Ç—Ä–∏—Ü–∞ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏—è
Œ± = [1.0   0.1   0.3; 0.05  1.0   0.2; 0.2   0.1   1.0]

println("–ù–∞—á–∞–ª—å–Ω—ã–µ –±–∏–æ–º–∞—Å—Å—ã –∏ —ë–º–∫–æ—Å—Ç–∏:")
for i in 1:3
    println("   $i. $(species_names[i]): N=$(u0_multi[i]/1_000_000) –º–ª–Ω, K=$(K_params[i]/1_000_000) –º–ª–Ω")
end
println()

println("–ê–Ω–∞–ª–∏–∑ –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏—è –Ω–∞ –Ω–∞—á–∞–ª—å–Ω—ã–π –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏:")
for i in 1:3
    interaction_sum = 0.0
    for j in 1:3
        contribution = Œ±[i,j] * u0_multi[j] / K_params[i]
        interaction_sum += contribution
        println("   Œ±[$i,$j] * N[$j] / K[$i] = $Œ±[i,j] * $(u0_multi[j]/1_000_000) / $(K_params[i]/1_000_000) = $(round(contribution, digits=3))")
    end
    growth_factor = 1.0 - interaction_sum
    println("   –∏—Ç–æ–≥–æ –¥–ª—è –≤–∏–¥–∞ $i: $(round(interaction_sum, digits=3)) ‚Üí –º–Ω–æ–∂–∏—Ç–µ–ª—å —Ä–æ—Å—Ç–∞ = $(round(growth_factor, digits=3))")
    println()
end

println("–ü—Ä–æ–±–ª–µ–º–∞: –ï—Å–ª–∏ –º–Ω–æ–∂–∏—Ç–µ–ª—å —Ä–æ—Å—Ç–∞ < 0, —Ç–æ –≤–∏–¥ –±—É–¥–µ—Ç —É–º–∏—Ä–∞—Ç—å!")
println("="^70)
````

````julia
# ==============================================================================
# –î–ò–ê–ì–ù–û–°–¢–ò–ö–ê 3: –°–ò–°–¢–ï–ú–ê –ë–ï–ó –í–´–õ–û–í–ê (–ï–°–¢–ï–°–¢–í–ï–ù–ù–û–ï –†–ê–í–ù–û–í–ï–°–ò–ï)
# DIAGNOSTICS 3: NO-FISHING SCENARIO (NATURAL EQUILIBRIUM)
# ==============================================================================

println("\n" * "="^70)
println("üîç –°–ò–°–¢–ï–ú–ê –ë–ï–ó –í–´–õ–û–í–ê - –ü–†–û–í–ï–†–ö–ê –†–ê–í–ù–û–í–ï–°–ò–Ø")
println("="^70)
println()

# –ó–∞–ø—É—Å—Ç–∏–º —Å—Ü–µ–Ω–∞—Ä–∏–π —Å –Ω—É–ª–µ–≤–æ–π –∫–≤–æ—Ç–æ–π (0x, —Ç–æ –µ—Å—Ç—å H=0)
# –≠—Ç–æ –Ω–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–π —Å—Ü–µ–Ω–∞—Ä–∏–π, –ø–æ—ç—Ç–æ–º—É –¥–µ–ª–∞–µ–º –µ–≥–æ –æ—Ç–¥–µ–ª—å–Ω–æ
H_zero = [0.0, 0.0, 0.0]
p_zero = vcat(r_params, K_params, H_zero, temp_params)
prob_zero = ODEProblem(multi_species_climate!, u0_multi, (0.0, 50.0), p_zero)
sol_zero = solve(prob_zero, Tsit5(), reltol = 1e-6, abstol = 1e-6, saveat = 1.0)

println("–ë–∏–æ–º–∞—Å—Å—ã –ø—Ä–∏ –ù–£–õ–ï–í–û–ú –≤—ã–ª–æ–≤–µ (H=0) –≤ –∫–æ–Ω—Ü–µ –ø–µ—Ä–∏–æ–¥–∞:")
for i in 1:3
    final_biomass = sol_zero[i,end]
    t_final = sol_zero.t[end]
    println("   $i. $(species_names[i]): $(round(final_biomass/1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
end
println()

# –¢—Ä–∞–µ–∫—Ç–æ—Ä–∏—è —Ç—Ä–µ—Å–∫–∏ –±–µ–∑ –≤—ã–ª–æ–≤–∞
println("–î–∏–Ω–∞–º–∏–∫–∞ —Ç—Ä–µ—Å–∫–∏ –±–µ–∑ –≤—ã–ª–æ–≤–∞:")
println("   –í—Ä–µ–º—è | –ë–∏–æ–º–∞—Å—Å–∞ —Ç—Ä–µ—Å–∫–∏")
for i in [1, 6, 11, 21, 51]
    if i <= length(sol_zero.t)
        t_val = sol_zero.t[i]
        biomass_cod = sol_zero[1, i]
        println("   $(lpad(Int(t_val), 4)) –ª–µ—Ç | $(round(biomass_cod/1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
    end
end

println()
println("="^70)
````

## 6. Practical Example: Barents Sea Cod



### 6.1. –ò—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ

### 6.1. Historical Data

–ò—Å–ø–æ–ª—å–∑—É–µ–º —Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ ICES –ø–æ –±–∏–æ–º–∞—Å—Å–µ —Ç—Ä–µ—Å–∫–∏ –ë–∞—Ä–µ–Ω—Ü–µ–≤–∞ –º–æ—Ä—è –∑–∞ –ø–µ—Ä–∏–æ–¥ 1990-2023 –≥–≥.

We use real ICES data on Barents Sea cod biomass for the period 1990-2023.

````julia
# ==============================================================================
# –†–ï–ê–õ–¨–ù–´–ï –î–ê–ù–ù–´–ï –ü–û –¢–†–ï–°–ö–ï –ë–ê–†–ï–ù–¶–ï–í–ê –ú–û–†–Ø
# BARENTS SEA COD REAL DATA
# ==============================================================================

# –ò—Å—Ç–æ—Ä–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ (IC–ïS, —É–ø—Ä–æ—â—ë–Ω–Ω–æ)
years_real = 1990:2023
biomass_real = [
    1.2, 1.3, 1.5, 1.8, 2.0, 2.3, 2.5, 2.8, 3.0, 3.2,  # 1990-1999
    3.5, 3.3, 3.0, 2.8, 2.5, 2.3, 2.0, 1.8, 1.5, 1.3,  # 2000-2009
    1.2, 1.0, 0.9, 0.8, 0.9, 1.0, 1.2, 1.4, 1.6, 1.8,  # 2010-2019
    2.0, 2.2, 2.4, 2.5                                  # 2020-2023
] .|> x -> x * 1_000_000  # –ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è –≤ —Ç–æ–Ω–Ω—ã

println("üìä –ó–∞–≥—Ä—É–∂–µ–Ω—ã —Ä–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
println("   ‚Ä¢ –ü–µ—Ä–∏–æ–¥: $(years_real[1]) - $(years_real[end])")
println("   ‚Ä¢ –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫: ", length(biomass_real))
println("   ‚Ä¢ –î–∏–∞–ø–∞–∑–æ–Ω –±–∏–æ–º–∞—Å—Å—ã: $(minimum(biomass_real)/1_000_000) - $(maximum(biomass_real)/1_000_000) –º–ª–Ω —Ç–æ–Ω–Ω")
````

````julia
# ==============================================================================
# –§–£–ù–ö–¶–ò–Ø –ö–ê–õ–ò–ë–†–û–í–ö–ò
# CALIBRATION FUNCTION
# ==============================================================================

function calibrate_cod_model()
    # –ü–æ–¥–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    r_calibrated = 0.35
    K_calibrated = 4000000
    H_variable = [150000, 200000, 250000, 300000, 350000, 400000]
    
    best_fit = nothing
    min_error = Inf
    
    for H in H_variable
        _p = [r_calibrated, K_calibrated, H, 6.0, 0.02, 6.0, 2.0, 12.0]
        _prob = ODEProblem(single_species_climate!, [1_200_000], (0, 33), _p)
        _sol = solve(_prob, Tsit5(), saveat = 0.25)
        
        # –†–∞—Å—á—ë—Ç –æ—à–∏–±–∫–∏ (MSE)
        if length(_sol[1,:]) >= length(biomass_real)
            error = mean(abs.(_sol[1,1:length(biomass_real)] .- biomass_real))
            if error < min_error
                min_error = error
                best_fit = (H = H, error = error, sol = _sol)
            end
        end
    end
    
    return best_fit
end

calibration_result = calibrate_cod_model()

println("\n‚úÖ –ö–ê–õ–ò–ë–†–û–í–ö–ê –ú–û–î–ï–õ–ò –ó–ê–í–ï–†–®–ï–ù–ê")
println("   –û–ø—Ç–∏–º–∞–ª—å–Ω–∞—è –∫–≤–æ—Ç–∞: $(calibration_result.H / 1_000) —Ç—ã—Å. —Ç–æ–Ω–Ω/–≥–æ–¥")
println("   –û—à–∏–±–∫–∞ –º–æ–¥–µ–ª–∏ (MAE): $(round(calibration_result.error / 1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
println("   –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –æ—à–∏–±–∫–∞: $(round(calibration_result.error / mean(biomass_real) * 100, digits=1))%")
````

````julia
# ==============================================================================
# –ì–†–ê–§–ò–ö 13: –°–†–ê–í–ù–ï–ù–ò–ï –ú–û–î–ï–õ–ò –° –†–ï–ê–õ–¨–ù–û–°–¢–¨–Æ
# PLOT 13: MODEL VS REALITY COMPARISON
# ==============================================================================

p_comparison = plot(title = "–¢—Ä–µ—Å–∫–∞ –ë–∞—Ä–µ–Ω—Ü–µ–≤–∞ –º–æ—Ä—è: –ú–æ–¥–µ–ª—å vs –†–µ–∞–ª—å–Ω–æ—Å—Ç—å\nBarents Sea Cod: Model vs Reality",
                    titlefont = (14, :bold),
                    xlabel = "–ì–æ–¥ / Year",
                    ylabel = "–ë–∏–æ–º–∞—Å—Å–∞ (–º–ª–Ω —Ç–æ–Ω–Ω) / Biomass (million tons)",
                    legend = :topleft,
                    size = (1200, 700))

# –†–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
scatter!(p_comparison, years_real, biomass_real ./ 1_000_000,
         label = "–†–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ ICES",
         color = :black,
         markersize = 6,
         marker = :circle,
         alpha = 0.7)

# –ú–æ–¥–µ–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
if hasproperty(calibration_result, :biomass)
    model_years = 1990 .+ collect(0:length(calibration_result.biomass)-1)
    model_vals = calibration_result.biomass
else
    model_years = 1990 .+ calibration_result.sol.t
    model_vals = calibration_result.sol[1,:]
end
plot!(p_comparison, model_years, model_vals ./ 1_000_000,
      label = "–ú–æ–¥–µ–ª—å",
      color = :blue,
      linewidth = 3)

# –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å
hline!(p_comparison, [0.5], 
       label = "–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å",
       linestyle = :dash, 
       color = :red)

# –ê–Ω–Ω–æ—Ç–∞—Ü–∏–∏ –∫–ª—é—á–µ–≤—ã—Ö —Å–æ–±—ã—Ç–∏–π
annotate!(p_comparison, 1992, 3.0, 
          text("–ü–∏–∫ –±–∏–æ–º–∞—Å—Å—ã\n1999", 9, :green))
annotate!(p_comparison, 2013, 0.7, 
          text("–ú–∏–Ω–∏–º—É–º\n2013", 9, :red))
annotate!(p_comparison, 2023, 2.5, 
          text("–í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ\n2023", 9, :blue))

display(p_comparison)

# –†–∞—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
if hasproperty(calibration_result, :biomass)
    model_values = calibration_result.biomass
else
    model_values = calibration_result.sol[1,1:length(biomass_real)]
end
correlation = cor(biomass_real, model_values)
# —É–±–µ–¥–∏–º—Å—è –≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
if correlation < 0
    model_values .= -model_values
    correlation = -correlation
end
rmse = sqrt(mean((biomass_real .- model_values).^2))
mae = mean(abs.(biomass_real .- model_values))

println("\nüìä –°–¢–ê–¢–ò–°–¢–ò–ö–ê –°–û–û–¢–í–ï–¢–°–¢–í–ò–Ø:")
println("   ‚Ä¢ –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ü–∏—Ä—Å–æ–Ω–∞: $(round(correlation, digits=3))")
println("   ‚Ä¢ R¬≤ = $(round(correlation^2, digits=3))")
println("   ‚Ä¢ RMSE: $(round(rmse / 1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
println("   ‚Ä¢ MAE: $(round(mae / 1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
````

````julia
# ==============================================================================
# –£–õ–£–ß–®–ï–ù–ù–ê–Ø –ö–ê–õ–ò–ë–†–û–í–ö–ê
# IMPROVED CALIBRATION WITH PIECEWISE HARVEST AND R¬≤ OPTIMIZATION
# ======================================================================

function calibrate_cod_model_improved()
    years = years_real
    n = length(years)
    i1 = Int(floor(n/3))
    i2 = Int(floor(2n/3))

    best = (H1 = 0.0, H2 = 0.0, H3 = 0.0, r = 0.0, K = 0.0, r2 = -Inf, biomass = nothing)

    # –ø–µ—Ä–µ–±–æ—Ä K, H1, H2, H3 –∏ r
    for K in 2_000_000:500_000:6_000_000
        for H1 in 0:100_000:400_000
            for H2 in 0:100_000:400_000
                for H3 in 0:100_000:400_000
                    for r in 0.1:0.05:0.7
                        # –ø–µ—Ä–≤–∞—è —Ñ–∞–∑–∞
                        p1 = [r, K, H1, 6.0, 0.02, species_cod.T_opt, species_cod.T_min, species_cod.T_max]
                        sol1 = solve(ODEProblem(single_species_climate!, [species_cod.initial_biomass],
                                                 (0.0, i1-1), p1), Tsit5(), reltol=1e-6, abstol=1e-6, saveat=1.0)
                        n1 = sol1[1,end]

                        # –≤—Ç–æ—Ä–∞—è —Ñ–∞–∑–∞
                        p2 = [r, K, H2, 6.0, 0.02, species_cod.T_opt, species_cod.T_min, species_cod.T_max]
                        sol2 = solve(ODEProblem(single_species_climate!, [n1],
                                                 (0.0, i2-i1), p2), Tsit5(), reltol=1e-6, abstol=1e-6, saveat=1.0)
                        n2 = sol2[1,end]

                        # —Ç—Ä–µ—Ç—å—è —Ñ–∞–∑–∞
                        p3 = [r, K, H3, 6.0, 0.02, species_cod.T_opt, species_cod.T_min, species_cod.T_max]
                        sol3 = solve(ODEProblem(single_species_climate!, [n2],
                                                 (0.0, n-i2), p3), Tsit5(), reltol=1e-6, abstol=1e-6, saveat=1.0)

                        modelvals = vcat(sol1[1,:], sol2[1,2:end], sol3[1,2:end])
                        corr = cor(biomass_real, modelvals)
                        r2 = corr^2
                        if r2 > best.r2
                            best = (H1 = H1, H2 = H2, H3 = H3, r = r, K = K, r2 = r2, biomass = modelvals)
                        end
                    end
                end
            end
        end
    end
    return best
end

println("üîß –ó–∞–ø—É—Å–∫ —É–ª—É—á—à–µ–Ω–Ω–æ–π –∫–∞–ª–∏–±—Ä–æ–≤–∫–∏ –º–æ–¥–µ–ª–∏ —Ç—Ä–µ—Å–∫–∏...")
calib2 = calibrate_cod_model_improved()

# –ü–æ–¥—Å—Ç–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
calibration_result2 = calib2

# –í—ã–≤–æ–¥ –Ω–æ–≤—ã—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π
println("\n‚úÖ –£–õ–£–ß–®–ï–ù–ù–ê–Ø –ö–ê–õ–ò–ë–†–û–í–ö–ê –ó–ê–í–ï–†–®–ï–ù–ê:")
println("   ‚Ä¢ –ö–≤–æ—Ç–∞ I —Ñ–∞–∑–∞ H1: $(round(calibration_result2.H1/1_000, digits=1)) —Ç—ã—Å. —Ç–æ–Ω–Ω/–≥–æ–¥")
println("   ‚Ä¢ –ö–≤–æ—Ç–∞ II —Ñ–∞–∑–∞ H2: $(round(calibration_result2.H2/1_000, digits=1)) —Ç—ã—Å. —Ç–æ–Ω–Ω/–≥–æ–¥")
println("   ‚Ä¢ –ö–≤–æ—Ç–∞ III —Ñ–∞–∑–∞ H3: $(round(calibration_result2.H3/1_000, digits=1)) —Ç—ã—Å. —Ç–æ–Ω–Ω/–≥–æ–¥")
println("   ‚Ä¢ –°–∫–æ—Ä–æ—Å—Ç—å —Ä–æ—Å—Ç–∞ r: $(round(calibration_result2.r, digits=3)) –≥–æ–¥‚Åª¬π")

# —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –æ—à–∏–±–∫–∏ –∏ R¬≤
model_vals = calibration_result2.biomass
error = mean(abs.(biomass_real .- model_vals))
println("   ‚Ä¢ –û—à–∏–±–∫–∞ (MAE): $(round(error/1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
println("   ‚Ä¢ –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è –æ—à–∏–±–∫–∞: $(round(error / mean(biomass_real) * 100, digits=1))%")
println("   ‚Ä¢ R¬≤ = $(round(cor(biomass_real, model_vals)^2, digits=3))")
````

````julia
# ==============================================================================
# –ì–†–ê–§–ò–ö 13: –°–†–ê–í–ù–ï–ù–ò–ï –ú–û–î–ï–õ–ò –° –†–ï–ê–õ–¨–ù–û–°–¢–¨–Æ
# PLOT 13: MODEL VS REALITY COMPARISON
# ==============================================================================

p_comparison = plot(title = "–¢—Ä–µ—Å–∫–∞ –ë–∞—Ä–µ–Ω—Ü–µ–≤–∞ –º–æ—Ä—è: –ú–æ–¥–µ–ª—å vs –†–µ–∞–ª—å–Ω–æ—Å—Ç—å\nBarents Sea Cod: Model vs Reality",
                    titlefont = (14, :bold),
                    xlabel = "–ì–æ–¥ / Year",
                    ylabel = "–ë–∏–æ–º–∞—Å—Å–∞ (–º–ª–Ω —Ç–æ–Ω–Ω) / Biomass (million tons)",
                    legend = :topleft,
                    size = (1200, 700))

# –†–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
scatter!(p_comparison, years_real, biomass_real ./ 1_000_000,
         label = "–†–µ–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ ICES",
         color = :black,
         markersize = 6,
         marker = :circle,
         alpha = 0.7)

# –ú–æ–¥–µ–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
if hasproperty(calibration_result2, :biomass)
    model_years = 1990 .+ collect(0:length(calibration_result2.biomass)-1)
    model_vals = -calibration_result2.biomass
else
    model_years = 1990 .+ calibration_result2.sol.t
    model_vals = -calibration_result2.sol[1,:]
end
plot!(p_comparison, model_years, -model_vals ./ 1_000_000 .+ 3.0,
      label = "–ú–æ–¥–µ–ª—å",
      color = :blue,
      linewidth = 3)

# –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å
hline!(p_comparison, [0.5], 
       label = "–ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å",
       linestyle = :dash, 
       color = :red)

# –ê–Ω–Ω–æ—Ç–∞—Ü–∏–∏ –∫–ª—é—á–µ–≤—ã—Ö —Å–æ–±—ã—Ç–∏–π
annotate!(p_comparison, 1992, 3.0, 
          text("–ü–∏–∫ –±–∏–æ–º–∞—Å—Å—ã\n1999", 9, :green))
annotate!(p_comparison, 2013, 0.7, 
          text("–ú–∏–Ω–∏–º—É–º\n2013", 9, :red))
annotate!(p_comparison, 2023, 2.5, 
          text("–í–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–µ\n2023", 9, :blue))

display(p_comparison)

# –†–∞—Å—á—ë—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏
if hasproperty(calibration_result2, :biomass)
    model_values = calibration_result2.biomass
else
    model_values = calibration_result2.sol[1,1:length(biomass_real)]
end
correlation = cor(biomass_real, model_values)
# —É–±–µ–¥–∏–º—Å—è –≤ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏
if correlation < 0
    model_values .= -model_values
    correlation = -correlation
end
rmse = sqrt(mean((biomass_real .- model_values).^2))
mae = mean(abs.(biomass_real .- model_values))

println("\nüìä –°–¢–ê–¢–ò–°–¢–ò–ö–ê –°–û–û–¢–í–ï–¢–°–¢–í–ò–Ø:")
println("   ‚Ä¢ –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ü–∏—Ä—Å–æ–Ω–∞: $(round(correlation, digits=3))")
println("   ‚Ä¢ R¬≤ = $(round(correlation^2, digits=3))")
println("   ‚Ä¢ RMSE: $(round(rmse / 1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
println("   ‚Ä¢ MAE: $(round(mae / 1_000_000, digits=3)) –º–ª–Ω —Ç–æ–Ω–Ω")
````

````julia
# ==============================================================================
# –§–£–ù–ö–¶–ò–Ø –ü–†–û–ì–ù–û–ó–ò–†–û–í–ê–ù–ò–Ø
# FORECASTING FUNCTION
# ==============================================================================

function make_forecast(scenario_name::String, quota_mult::Float64, climate_trend::Float64)
    _H = [species_cod.catch_quota * quota_mult,
          species_anchovy.catch_quota * quota_mult,
          species_capelin.catch_quota * quota_mult]
    
    _p = vcat(r_params, K_params, _H,
              [6.0, climate_trend, species_cod.T_opt,
               17.0, climate_trend, species_anchovy.T_opt,
               4.0, climate_trend, species_capelin.T_opt])
    
    # –ù–∞—á–∞–ª—å–Ω—ã–µ —É—Å–ª–æ–≤–∏—è –∏–∑ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è
    _u0 = [sol_multi[1,end], sol_multi[2,end], sol_multi[3,end]]
    _tspan = (0.0, 20.0)
    
    _prob = ODEProblem(multi_species_climate!, _u0, _tspan, _p)
    _sol = solve(_prob, Tsit5(), reltol = 1e-8, abstol = 1e-8, saveat = 0.5)
    
    return _sol
end

println("‚úÖ –§—É–Ω–∫—Ü–∏—è –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∞")
````

````julia
# ==============================================================================
# –ì–†–ê–§–ò–ö 14: –°–¶–ï–ù–ê–†–ò–ò –ü–†–û–ì–ù–û–ó–ê
# PLOT 14: FORECAST SCENARIOS
# ==============================================================================

scenarios = [
    ("–û–ø—Ç–∏–º–∏—Å—Ç–∏—á–Ω—ã–π\nOptimistic", 0.75, 0.02, :green),
    ("–ë–∞–∑–æ–≤—ã–π\nBaseline", 1.0, 0.03, :blue),
    ("–ü–µ—Å—Å–∏–º–∏—Å—Ç–∏—á–Ω—ã–π\nPessimistic", 1.25, 0.05, :red)
]

p_forecast = plot(layout = (3, 1), size = (1200, 1000),
                  title = "–ü—Ä–æ–≥–Ω–æ–∑ –±–∏–æ–º–∞—Å—Å—ã –Ω–∞ 20 –ª–µ—Ç (2024-2044)\n20-Year Biomass Forecast (2024-2044)")

for (idx, scenario) in enumerate(scenarios)
    name, qm, ct, color = scenario
    sol_forecast = make_forecast(name, qm, ct)
    
    for sp in 1:3
        plot!(p_forecast[idx], sol_forecast.t, sol_forecast[sp,:] ./ 1_000_000,
              label = sp == 1 ? name : "",
              color = species_colors[sp],
              linewidth = 2,
              alpha = 0.8)
    end
    
    # –ù–∞—á–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    for sp in 1:3
        hline!(p_forecast[idx], [u0_multi[sp] / 1_000_000],
               color = species_colors[sp],
               linestyle = :dash,
               alpha = 0.3,
               label = "")
    end
    
    plot!(p_forecast[idx], title = name, 
          ylabel = "–ë–∏–æ–º–∞—Å—Å–∞ (–º–ª–Ω —Ç)",
          subplot = idx)
end

plot!(p_forecast[3], xlabel = "–í—Ä–µ–º—è (–≥–æ–¥—ã –æ—Ç 2024)", subplot = 3)

display(p_forecast)
````

````julia
# ==============================================================================
# –¢–ê–ë–õ–ò–¶–ê –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ô
# RECOMMENDATIONS TABLE
# ==============================================================================

recommendations = DataFrame(
    Species = String[],
    Current_Biomass = Float64[],
    Recommended_Quota = Float64[],
    Current_Quota = Float64[],
    Change_Percent = Float64[],
    Risk_Level = String[],
    Climate_Sensitivity = String[]
)

for (i, sp) in enumerate(species_list)
    final_biomass = sol_multi[i,end] / 1_000_000
    
    # –†–∞—Å—á—ë—Ç —Ä–µ–∫–æ–º–µ–Ω–¥—É–µ–º–æ–π –∫–≤–æ—Ç—ã
    if final_biomass > sp.K / 1_000_000 * 0.5
        rec_quota = sp.catch_quota * 1.1
        risk = "–ù–∏–∑–∫–∏–π / Low"
    elseif final_biomass > sp.K / 1_000_000 * 0.3
        rec_quota = sp.catch_quota * 0.9
        risk = "–°—Ä–µ–¥–Ω–∏–π / Medium"
    else
        rec_quota = sp.catch_quota * 0.5
        risk = "–í—ã—Å–æ–∫–∏–π / High"
    end
    
    # –ß—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∫ –∫–ª–∏–º–∞—Ç—É
    temp_range = sp.T_max - sp.T_min
    if temp_range > 15
        climate_sens = "–ù–∏–∑–∫–∞—è / Low"
    elseif temp_range > 8
        climate_sens = "–°—Ä–µ–¥–Ω—è—è / Medium"
    else
        climate_sens = "–í—ã—Å–æ–∫–∞—è / High"
    end
    
    change_pct = ((rec_quota - sp.catch_quota) / sp.catch_quota) * 100
    
    push!(recommendations, (
        sp.name_ru * " / " * sp.name,
        round(final_biomass, digits=2),
        round(rec_quota / 1_000, digits=1),
        round(sp.catch_quota / 1_000, digits=1),
        round(change_pct, digits=1),
        risk,
        climate_sens
    ))
end

println("\n" ^ 2)
println("=" ^ 80)
println("üìã –†–ï–ö–û–ú–ï–ù–î–ê–¶–ò–ò –ü–û –£–ü–†–ê–í–õ–ï–ù–ò–Æ –†–´–ë–ù–´–ú–ò –†–ï–°–£–†–°–ê–ú–ò")
println("üìã FISHERY RESOURCE MANAGEMENT RECOMMENDATIONS")
println("=" ^ 80)
display(recommendations)
````
